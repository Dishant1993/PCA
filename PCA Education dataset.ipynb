This code appears to be analyzing a DataFrame, df, with continuous columns by iterating through each column, printing the skewness of each column, and then creating two plots for each column:

A histogram of the column's values with vertical lines indicating the mean and median.
A box plot of the column's values.
However, there is an issue with the code. The subplot function is being used with a 3-panel grid, but the third panel should have the position (1,2) instead of (1,3).

Here is the corrected code with comments for clarification:

```python
cont_cols = list(df.columns)

for col in cont_cols:
    # Print the skewness of the column
    print(col)
    print('Skew :',np.round(df[col].skew(),2))
    
    # Create a figure with a 2-panel grid
    plt.figure(figsize=(25,6))
    
    # Add the histogram subplot to the first panel
    plt.subplot(1,2,1)
    sns.distplot(df[col],norm_hist=False,kde=False,bins=50,hist_kws=dict(edgecolor="black",linewidth=1.5))
    plt.vlines(df[col].mean(),ymin=0,ymax=40,color='red', linewidth=3)
    plt.vlines(df[col].median(),ymin=0,ymax=40,color='green', linewidth=3)
    plt.ylabel('count')
    
    # Add the box plot subplot to the second panel
    plt.subplot(1,2,2)
    sns.boxplot(df[col])
    
    # Show the plot
    plt.show()

    
    
```


This code creates a heatmap of the correlation matrix of df. The np.triu function creates a matrix of True values above the diagonal and False values below the diagonal. This is used to create a mask to hide the lower triangle of the heatmap (which is redundant due to the symmetry of the correlation matrix). The annot=True argument adds numeric values to each cell of the heatmap. The cmap argument sets the color scheme of the heatmap.

```python
plt.figure(figsize=(25,6))
mask = np.triu(np.ones_like(df.corr(),dtype=bool))
sns.heatmap(df.corr(),annot=True,mask=mask,cmap='YlGnBu')
```

In this code, pc_cols is a list of the column names for the principal components. cov_matrix is the covariance matrix of the standardized data. eig_vals and eig_vecs are the eigenvalues and eigenvectors of the covariance matrix, respectively. sorted_indices are the indices that sort the eigenvalues in descending order. pcs is the matrix of principal components, obtained by multiplying the standardized data matrix (df.values) with the eigenvectors. Finally, pc_df_numpy is a DataFrame that contains the principal components with rounded values to 2 decimal places. The head(10).T method call transposes the first 10 principal components for easier viewing.

```python
#Calculation of pc components manually using numpy module
pc_cols = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15','PC16','PC17']
cov_matrix = np.cov(df.T)
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
pcs = eig_vecs.T
pc_df_numpy = pd.DataFrame(np.round(pcs, 2),index=pc_cols, columns=cont_cols)
pc_df_numpy.head(10).T
```

This code calculates the percentage of variance explained by each principal component, and also calculates the cumulative explained variance. The sorted_eig_vals variable should contain the eigenvalues in descending order, which can be obtained by sorting the eig_vals variable that you computed earlier.

```python
sorted_eig_vals = sorted(eig_vals, reverse=True)

tot=np.sum(sorted_eig_vals)
var_exp=[(i/tot)*100 for i in sorted_eig_vals]
cum_var_exp=np.cumsum(var_exp)
print('Explained variance by each principal component : \n',np.round(var_exp,2))
```

Here, cum_var_exp is a numpy array containing the cumulative explained variance of each principal component. The np.round() function is used to round the values in the array to two decimal places for better readability.

```python
print('Cummalative explained variance : \n',np.round(cum_var_exp,2))
```

This will display the plot showing the cumulative variance explained by each additional principal component.

```python
plt.figure(figsize=(18, 6))
plt.plot(cum_var_exp,marker='o')
plt.xticks(np.arange(0,17),labels=np.arange(1,18))
plt.xlabel('no of principal components')
plt.ylabel('no of cumulative variance explained')
```
